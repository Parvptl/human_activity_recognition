**HANDLING OUTLIERS**

```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("Dataset shape (rows, cols):", train_df.shape)
total_rows = len(train_df)

# If your label column is named 'Activity', exclude it from numeric checks
label_col = 'Activity' if 'Activity' in train_df.columns else None
X = train_df.drop(columns=[label_col]) if label_col else train_df.copy()

# thresholds / heuristics
z_threshold = 3.0
skew_cutoff = 1.0
kurtosis_excess_cutoff = 2.0

outlier_info = {}
summary_rows = []

for feat in X.columns:
    col = X[feat].dropna()
    # robustly compute skew/kurtosis
    s_skew = float(col.skew())
    s_kurt_excess = float(col.kurtosis())
    
    # choose detection method
    if (abs(s_skew) < skew_cutoff) and (s_kurt_excess < kurtosis_excess_cutoff):
        # likely near-normal -> use Z-score
        mu = col.mean()
        sigma = col.std(ddof=0) + 1e-12
        z = (col - mu) / sigma
        mask = z.abs() > z_threshold
        method = 'Z-score'
    else:
        # skewed -> use IQR
        Q1 = col.quantile(0.25)
        Q3 = col.quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        mask = (col < lower) | (col > upper)
        method = 'IQR'
    
    indices = col.index[mask.fillna(False)].tolist()
    outlier_info[feat] = {'method': method, 'indices': indices}
    summary_rows.append({
        'feature': feat,
        'method': method,
        'outlier_count': len(indices),
        'skewness': s_skew,
        'kurtosis_excess': s_kurt_excess
    })

# summary table (per-feature)
summary_df = pd.DataFrame(summary_rows).sort_values('outlier_count', ascending=False)
display(summary_df.head(30))

# save summary for record (in working dir)
summary_csv_path = "../data/outlier_summary.csv"
summary_df.to_csv(summary_csv_path, index=False)
print(f"Saved per-feature outlier summary to: {summary_csv_path}")

# totals
total_outliers_sum = summary_df['outlier_count'].sum()   # sum across features (may double-count rows)
unique_outlier_indices = set()
for info in outlier_info.values():
    unique_outlier_indices.update(info['indices'])
unique_outlier_count = len(unique_outlier_indices)

print("\n--- Outlier counts ---")
print("Total rows in dataset:", total_rows)
print("Sum of outlier counts across features (may double-count rows):", total_outliers_sum)
print(f"Unique rows flagged as outlier in any feature: {unique_outlier_count} "
      f"({(unique_outlier_count/total_rows)*100:.4f}% of rows)")

# Decision threshold (you can change 0.5% as you prefer)
significance_threshold_pct = 0.5
if unique_outlier_count == 0 or (unique_outlier_count/total_rows)*100 < significance_threshold_pct:
    print(f"\nConclusion: No significant outliers found (unique rows < {significance_threshold_pct}% of total). ")
else:
    print(f"\nConclusion: {unique_outlier_count} rows ({(unique_outlier_count/total_rows)*100:.3f}%) flagged — consider inspecting/removing.")

# Optional: show some example rows that were flagged
if unique_outlier_count > 0:
    sample_idxs = list(unique_outlier_indices)[:5]
    print("\nExample flagged rows (first 5 indices):", sample_idxs)
    display(train_df.loc[sample_idxs])

# Visual diagnostic: distribution of per-row outlier-counts
row_outlier_counts = pd.Series(0, index=train_df.index, dtype=int)
for feat, info in outlier_info.items():
    row_outlier_counts.loc[info['indices']] += 1

plt.figure(figsize=(8,4))
sns.histplot(row_outlier_counts, bins=30, kde=False)
plt.title("Histogram of outlier-feature-count per row (how many features flagged per row)")
plt.xlabel("Number of features flagged as outlier for a row")
plt.ylabel("Row count")
plt.tight_layout()
plt.show()
```

During initial exploration, a univariate Z-score/IQR-based method flagged nearly 92% of samples as outliers.
While this may look alarming, such a high ratio is not truly meaningful in a high-dimensional sensor dataset like UCI HAR.
The reason lies in how univariate detection behaves:

Each feature is treated independently.
With over 500 correlated sensor features, even small deviations in a few dimensions make almost every sample appear “extreme.”

Natural motion variability.
Human activity signals (acceleration, gyroscope) inherently vary between subjects and activities.
These variations are real patterns, not data errors.

Feature correlation ignored.
Z-score/IQR rules assume independent, normally distributed features—an invalid assumption for sensor fusion data.

Therefore, removing those “outliers” would risk discarding valid human activity information and severely harming model generalization.
To assess data quality properly, we switched to a multivariate anomaly-detection approach (PCA + IsolationForest), which considers all features jointly.

```python
# ---------------- MULTIVARIATE OUTLIER ANALYSIS ---------------- #
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

print("\nPerforming multivariate (PCA + IsolationForest) outlier analysis...")

# Drop label column if present
label_col = 'Activity' if 'Activity' in train_df.columns else None
X = train_df.drop(columns=[label_col]) if label_col else train_df.copy()

#  Standardize features (important for PCA / IsolationForest)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reduce to 2 principal components for visualization
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

# Fit IsolationForest
iso = IsolationForest(contamination=0.01, random_state=42)
outlier_pred = iso.fit_predict(X_scaled)  # -1 = outlier, 1 = inlier

# Calculate fraction
outlier_fraction = (outlier_pred == -1).mean() * 100
print(f"\nEstimated outlier fraction (multivariate): {outlier_fraction:.3f}%")

# Visualize in PCA space
plt.figure(figsize=(10,7))
sns.scatterplot(
    x=X_pca[:,0],
    y=X_pca[:,1],
    hue=outlier_pred,
    palette={1: "skyblue", -1: "red"},
    alpha=0.7,
    s=25
)
plt.title("PCA Visualization with IsolationForest Outlier Detection", fontsize=14)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(["Inlier", "Outlier"], loc="best")
plt.tight_layout()
plt.show()

# Final interpretation
if outlier_fraction < 1:
    print(f"No significant outliers detected ({outlier_fraction:.3f}% flagged). Dataset is clean ")
else:
    print(f"Some potential anomalies detected ({outlier_fraction:.2f}%). Consider investigating visually.")
```

After applying PCA and IsolationForest, we observed that:

The data points form a continuous, compact cluster in the 2D PCA space, with no visibly isolated points.

The estimated outlier fraction was less than 1%, which is within a normal tolerance for natural variance in sensor readings.

Most detected “outliers” are likely due to minor motion differences or noise, not actual anomalies or recording errors.

**Conclusion:**
The multivariate analysis confirms that the UCI HAR dataset is well-cleaned and consistent.
There are no significant outliers, and all data points can safely be used for model training without any removal or modification.

**We are not performing outlier detection on the test data because doing so would cause data leakage. The outlier detection rules would be "learned" using information from the test set, giving us an unfairly optimistic and unrealistic measure of the model's real-world performance.**

**ENCODING CATEGORICAL DATA**

```python
# ---------------- LABEL ENCODING ---------------- #
from sklearn.preprocessing import LabelEncoder

# Check for categorical columns
categorical_cols = train_df.select_dtypes(include=['object']).columns.tolist()
print("Categorical columns found:", categorical_cols)

# Encode each categorical column (if any)
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    train_df[col] = le.fit_transform(train_df[col])
    test_df[col] = le.transform(test_df[col])
    label_encoders[col] = le
    print(f"Encoded '{col}' -> {len(le.classes_)} unique categories")

# Verify encoding
display(train_df.head())
print('------------------------------------------------------------')
display(test_df.head())

# show label mapping for target
if 'Activity' in label_encoders:
    mapping = dict(zip(label_encoders['Activity'].classes_, label_encoders['Activity'].transform(label_encoders['Activity'].classes_)))
    print("\nActivity label mapping:")
    for k, v in mapping.items():
        print(f"{k:>12}  -->  {v}")
```

```python
#Analyze the Target Variable ('Activity')
print("Activity Distribution")
activity_counts = train_df['Activity'].value_counts()
display(activity_counts)

# --- 4. Visualize the Activity Distribution ---
plt.figure(figsize=(12, 7))
ax = sns.countplot(x='Activity', data=train_df, order=activity_counts.index, palette='viridis')
plt.title('Activity Counts', fontsize=16)
plt.ylabel('Number of Samples', fontsize=12)
plt.xlabel('Activity', fontsize=12)
plt.xticks(rotation=45)

# Calculate and add percentage labels to each bar
total_samples = len(train_df)
for p in ax.patches:
    percentage = f'{(p.get_height() / total_samples) * 100:.1f}%'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.show()

print("\n--------------------------------------------------------------------------------------------------------------------------------------------")
plt.figure(figsize=(10, 8))
activity_percentages = train_df['Activity'].value_counts(normalize=True) * 100
plt.pie(activity_percentages, labels=activity_percentages.index, autopct='%1.1f%%', startangle=90,
        wedgeprops=dict(width=0.4), colors=sns.color_palette('viridis', 6))
plt.title('Proportional Distribution of Activities', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

1. The visualizations confirm that the dataset is very well-balanced.
2. Each activity constitutes between 13% and 19% of the total samples, with no single activity dominating the dataset.
3. This balance is ideal for training a robust classification model without needing resampling techniques like SMOTE.
