# **Regression with MET Values**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import cross_val_score

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
```

**LOAD DATA and ASSIGN MET VALUES (CONTINOUS) TO THE TARGET**

```python
train_df = pd.read_csv('../data/cleaned_train.csv')
test_df = pd.read_csv('../data/cleaned_test.csv')

print(f"Training data shape: {train_df.shape}")
print(f"Test data shape: {test_df.shape}")

# MET values for each activity
MET_VALUES = {
    0: 1.0,   # LAYING
    1: 1.3,   # SITTING
    2: 1.4,   # STANDING
    3: 3.0,   # WALKING
    4: 3.5,   # WALKING_DOWNSTAIRS
    5: 8.0    # WALKING_UPSTAIRS
}

# Map activities to MET values
train_df['MET'] = train_df['Activity'].map(MET_VALUES)
test_df['MET'] = test_df['Activity'].map(MET_VALUES)

print("\nMET Value Mapping:")
print("  LAYING → 1.0")
print("  SITTING → 1.3")
print("  STANDING → 1.4")
print("  WALKING → 3.0")
print("  WALKING_DOWNSTAIRS → 3.5")
print("  WALKING_UPSTAIRS → 8.0")
```

**DATA PREPARATION**

```python
X_train = train_df.drop(['subject', 'Activity', 'MET'], axis=1)
y_train = train_df['MET']

X_test = test_df.drop(['subject', 'Activity', 'MET'], axis=1)
y_test = test_df['MET']

print(f"Number of features: {X_train.shape[1]}")
print(f"Number of training samples: {X_train.shape[0]}")
print(f"Number of test samples: {X_test.shape[0]}")

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\n Visualizing MET distribution...")

plt.figure(figsize=(10, 5))
plt.hist(y_train, bins=20, edgecolor='black', alpha=0.7, color='skyblue')
plt.xlabel('MET Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of MET Values in Training Set', fontsize=14)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

**TRAIN REGRESSION MODELS**

```python
results = {}

# --- Linear Regression ---
print("\n[Model 1] Linear Regression")
lr_model = LinearRegression()
lr_model.fit(X_train_scaled, y_train)
y_pred_lr = lr_model.predict(X_test_scaled)

lr_rmse = np.sqrt(mean_squared_error(y_test, y_pred_lr))
lr_mae = mean_absolute_error(y_test, y_pred_lr)
lr_r2 = r2_score(y_test, y_pred_lr)

print(f"  RMSE: {lr_rmse:.4f}")
print(f"  MAE:  {lr_mae:.4f}")
print(f"  R²:   {lr_r2:.4f}")

results['Linear Regression'] = {
    'RMSE': lr_rmse, 'MAE': lr_mae, 'R2': lr_r2, 'predictions': y_pred_lr
}

# --- Ridge Regression ---
print("\n[Model 2] Ridge Regression")
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train_scaled, y_train)
y_pred_ridge = ridge_model.predict(X_test_scaled)

ridge_rmse = np.sqrt(mean_squared_error(y_test, y_pred_ridge))
ridge_mae = mean_absolute_error(y_test, y_pred_ridge)
ridge_r2 = r2_score(y_test, y_pred_ridge)

print(f"  RMSE: {ridge_rmse:.4f}")
print(f"  MAE:  {ridge_mae:.4f}")
print(f"  R²:   {ridge_r2:.4f}")

results['Ridge Regression'] = {
    'RMSE': ridge_rmse, 'MAE': ridge_mae, 'R2': ridge_r2, 'predictions': y_pred_ridge
}

# --- Model 3: Decision Tree ---
print("\n[Model 3] Decision Tree Regressor")
dt_model = DecisionTreeRegressor(max_depth=10, random_state=42)
dt_model.fit(X_train_scaled, y_train)
y_pred_dt = dt_model.predict(X_test_scaled)

dt_rmse = np.sqrt(mean_squared_error(y_test, y_pred_dt))
dt_mae = mean_absolute_error(y_test, y_pred_dt)
dt_r2 = r2_score(y_test, y_pred_dt)

print(f"  RMSE: {dt_rmse:.4f}")
print(f"  MAE:  {dt_mae:.4f}")
print(f"  R²:   {dt_r2:.4f}")

results['Decision Tree'] = {
    'RMSE': dt_rmse, 'MAE': dt_mae, 'R2': dt_r2, 'predictions': y_pred_dt
}

# --- Model 4: Random Forest ---
print("\n[Model 4] Random Forest Regressor")
rf_model = RandomForestRegressor(n_estimators=100, max_depth=20, random_state=42)
rf_model.fit(X_train_scaled, y_train)
y_pred_rf = rf_model.predict(X_test_scaled)

rf_rmse = np.sqrt(mean_squared_error(y_test, y_pred_rf))
rf_mae = mean_absolute_error(y_test, y_pred_rf)
rf_r2 = r2_score(y_test, y_pred_rf)

print(f"  RMSE: {rf_rmse:.4f}")
print(f"  MAE:  {rf_mae:.4f}")
print(f"  R²:   {rf_r2:.4f}")

results['Random Forest'] = {
    'RMSE': rf_rmse, 'MAE': rf_mae, 'R2': rf_r2, 'predictions': y_pred_rf
}

# --- Model 5: Support Vector Regressor ---
print("\n[Model 5] Support Vector Regressor (SVR)")
svr_model = SVR(kernel='rbf')
svr_model.fit(X_train_scaled, y_train)
y_pred_svr = svr_model.predict(X_test_scaled)

svr_rmse = np.sqrt(mean_squared_error(y_test, y_pred_svr))
svr_mae = mean_absolute_error(y_test, y_pred_svr)
svr_r2 = r2_score(y_test, y_pred_svr)

print(f"  RMSE: {svr_rmse:.4f}")
print(f"  MAE:  {svr_mae:.4f}")
print(f"  R²:   {svr_r2:.4f}")

results['SVR'] = {
    'RMSE': svr_rmse, 'MAE': svr_mae, 'R2': svr_r2, 'predictions': y_pred_svr
}
```

**VISUALIZE RESULTS**

```python
comparison_df = pd.DataFrame({
    'Model': list(results.keys()),
    'RMSE': [results[m]['RMSE'] for m in results.keys()],
    'MAE': [results[m]['MAE'] for m in results.keys()],
    'R² Score': [results[m]['R2'] for m in results.keys()]
})

comparison_df = comparison_df.sort_values('R² Score', ascending=False)
print("\n", comparison_df.to_string(index=False))

# Model comparison bar chart
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# RMSE
axes[0].barh(comparison_df['Model'], comparison_df['RMSE'], color='coral')
axes[0].set_xlabel('RMSE', fontsize=11)
axes[0].set_title('Root Mean Squared Error\n(Lower is Better)', fontsize=12)
axes[0].invert_yaxis()

# MAE
axes[1].barh(comparison_df['Model'], comparison_df['MAE'], color='skyblue')
axes[1].set_xlabel('MAE', fontsize=11)
axes[1].set_title('Mean Absolute Error\n(Lower is Better)', fontsize=12)
axes[1].invert_yaxis()

# R² Score
axes[2].barh(comparison_df['Model'], comparison_df['R² Score'], color='lightgreen')
axes[2].set_xlabel('R² Score', fontsize=11)
axes[2].set_title('R² Score\n(Higher is Better)', fontsize=12)
axes[2].invert_yaxis()

plt.tight_layout()
plt.show()
```

**BEST MODEL ANALYSIS**

```python
best_model_name = comparison_df.iloc[0]['Model']
best_predictions = results[best_model_name]['predictions']

print(f"\nBest Model: {best_model_name}")
print(f"  RMSE: {results[best_model_name]['RMSE']:.4f}")
print(f"  MAE:  {results[best_model_name]['MAE']:.4f}")
print(f"  R²:   {results[best_model_name]['R2']:.4f}")

# Actual vs Predicted plot
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Scatter plot
axes[0].scatter(y_test, best_predictions, alpha=0.6, edgecolors='k', linewidth=0.5)
axes[0].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 
             'r--', lw=2, label='Perfect Prediction')
axes[0].set_xlabel('Actual MET Values', fontsize=12)
axes[0].set_ylabel('Predicted MET Values', fontsize=12)
axes[0].set_title(f'Actual vs Predicted - {best_model_name}', fontsize=14)
axes[0].legend()
axes[0].grid(alpha=0.3)

# Residual plot
residuals = y_test - best_predictions
axes[1].scatter(best_predictions, residuals, alpha=0.6, edgecolors='k', linewidth=0.5)
axes[1].axhline(y=0, color='r', linestyle='--', lw=2)
axes[1].set_xlabel('Predicted MET Values', fontsize=12)
axes[1].set_ylabel('Residuals (Actual - Predicted)', fontsize=12)
axes[1].set_title(f'Residual Plot - {best_model_name}', fontsize=14)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

**SUMMARY OF WORK**

### MET Value Regression Analysis

#### Objective
Predict Metabolic Equivalent of Task (MET) values from accelerometer sensor data across six physical activities using regression models.

#### MET Value Assignment
Activities mapped to continuous energy expenditure values:
- LAYING → 1.0
- SITTING → 1.3
- STANDING → 1.4
- WALKING → 3.0
- WALKING_DOWNSTAIRS → 3.5
- WALKING_UPSTAIRS → 8.0

#### Models Evaluated
Five regression algorithms trained on standardized sensor features:
- Linear Regression
- Ridge Regression (α=1.0)
- Decision Tree Regressor (max_depth=10)
- Random Forest Regressor (100 estimators, max_depth=20)
- Support Vector Regressor (RBF kernel)

#### Evaluation Metrics
Models assessed using:
- Root Mean Squared Error (RMSE)
- Mean Absolute Error (MAE)
- R² Score (coefficient of determination)

#### Key Visualizations

##### Distribution Analysis
Histogram showing MET value frequency distribution in training data, revealing the six discrete activity intensity levels.

##### Model Comparison
Three-panel bar chart comparing all models across RMSE, MAE, and R² metrics, sorted by R² performance.

##### Best Model Diagnostics
Two-panel analysis of top performer:
- Scatter plot of actual vs predicted MET values with perfect prediction line
- Residual plot showing prediction errors across MET value range

#### Expected Outcomes
The analysis identifies which regression approach best captures the relationship between sensor measurements and energy expenditure, with Random Forest and SVR typically excelling due to their ability to model non-linear patterns in activity data.
